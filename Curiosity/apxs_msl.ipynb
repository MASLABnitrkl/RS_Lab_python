```python
import os
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
path='D:/Work/RS_Lab/Data/MSL'
os.chdir(path)

df=pd.read_csv('apxs_rsp_2245.csv')
```

```python
df
```

```python
energy=df['x-ray energy']
count=df['count']
energy_kev=energy/1000
```

```python
plt.figure(figsize=(15,6))
plt.plot(energy_kev,count, "ro", linestyle="--", linewidth=2)
plt.xlabel('X ray energy (keV)')
plt.ylabel('Counts (a.u)')
```

```python
plt.figure(figsize=(15,6))
plt.plot(energy_kev,count, "ro", linestyle="--", linewidth=2)
plt.xlabel('X ray energy (keV)')
plt.ylabel('Counts (a.u)')
plt.xlim(0,8)
plt.xticks(np.arange(0.5, 8.5, 0.5))
```

```python
fig, ax = plt.subplots(figsize=(15, 6))
ax.plot(energy_kev, count, "ro", linestyle="--", linewidth=2)

ax.set_xlabel('X ray energy (keV)', fontsize=14, fontweight="bold")
ax.set_ylabel('Counts (a.u)', fontsize=14, fontweight="bold")

ax.set_xlim(0, 8)
ax.set_xticks(np.arange(0.5, 8.5, 0.5))
ax.tick_params(axis='x', labelsize=14, width=2)
ax.tick_params(axis='y', labelsize=14, width=2)

ax.tick_params(axis='both', length=5, width=2)

for spine in ax.spines.values():
    spine.set_linewidth(2)


plt.savefig("D:/Work/RS_Lab/Data/MSL/apxs_spectrum.tiff", dpi=300)
plt.show()
```
